## 10.
- Open the METABRIC data. Make a KDE of 'Overall Survival (Months)'.
- Let survival time, $T$ be distributed exponentially with parameter $\lambda$. Then its distribution is
$$
F(t) = 1 - e^{-\lambda t} = p[T\le t].
$$
- Derive the density for this distribution.
- Write out the likelihood.
- Maximize the likelihood with respect to $\lambda$. What is the MLE, $\hat{\lambda}$?
- Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
- Bootstrap the sampling density/distribution of $\hat{\lambda}$.

```python
import scipy as sp
```

```python
mb.head(3).T
```

```python
sns.kdeplot(mb['Overall Survival (Months)'])
```

```python
y = mb['Overall Survival (Months)']

print('MLE', 1/y.mean())
```

```python
mu_hat = y.mean()

grid = np.linspace(y.min(), y.max(), 100)

pdf = sp.stats.expon.pdf(grid, scale=mu_hat)  # loc=0 by default
cdf = sp.stats.expon.cdf(grid, scale=mu_hat)

fig, axes = plt.subplots(1,2,figsize=(14,5))

# Left: KDE + PDF
sns.kdeplot(x=y, ax=axes[0], label='KDE')
axes[0].plot(grid, pdf, linestyle='--', color='red', label='Exponential PDF (MLE)')
axes[0].set_title('KDE vs Exponential PDF')
axes[0].legend()

# Right: ECDF + CDF
sns.ecdfplot(x=y, ax=axes[1], label='ECDF')
axes[1].plot(grid, cdf, linestyle='--', color='red', label='Exponential CDF (MLE)')
axes[1].set_title('ECDF vs Exponential CDF')
axes[1].legend()

plt.show()
```

This fit is far from perfect. The ECDF seems like it fits pretty well but the KDE seems very far off. That being said it fits well enough to work in my opinion.

```python
# Estimation:
def mle(y):
    mu_hat = np.mean(y)
    lambda_hat = 1/np.mean(y)
    return mu_hat, lambda_hat

# Bootstrapping:
S = 5000
sample = [ mle(y.sample(frac=1.0,replace=True)) for s in range(S)]

# Sampling distributions:
mu_hat_sample = [ row[0] for row in sample]
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=mu_hat_sample,ax=axes[0],label='kde')
axes[0].set_title('Sampling Density, $\\hat{\\mu}$')
sns.ecdfplot(x=mu_hat_sample,ax=axes[1],label='ecdf')
axes[1].set_title('Sampling Distribution, $\\hat{\\mu}$')
plt.show()

lambda_hat_samples = [ row[1] for row in sample]
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=lambda_hat_samples,ax=axes[0],label='kde')
axes[0].set_title('Sampling Density, $\\hat{\\lambda}$')
sns.ecdfplot(x=lambda_hat_samples,ax=axes[1],label='ecdf')
axes[1].set_title('Sampling Distribution, $\\hat{\\lambda}$')
plt.show()
```

- Open the METABRIC data. Make a histogram of 'Mutation Count' with around 50 bins. Let $Y$ be the mutation count the random variable, and $y_i$ the mutation count for patient $i$.
- We're going to model this using the Poisson Distribution. The probability that $y_i = k$ is given by

pr[y_i = k ] = \frac{ \lambda^k e^{-\lambda} }{k!}
$$
- Write out the likelihood.

```python
mb.head(3).T
```

```python
sns.histplot(mb['Mutation Count'], bins=50)
```

```python
y = mb['Mutation Count']
mu_hat = np.mean(y)

grid = np.arange(min(y), max(y)+1)  

pmf = sp.stats.poisson.pmf(grid, mu=mu_hat)
cdf = sp.stats.poisson.cdf(grid, mu=mu_hat)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=y,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=pmf,ax=axes[0],label='pmf',linestyle='--')
axes[0].set_title('KDE')
sns.ecdfplot(x=y,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('ECDF')
plt.show()
```

This seems to do a pretty good job of apporximating the actual mutation counts. We see the KDE and ECDF are relatively tight compared to the pmf and cdf.

```python
# Estimation:
def mle(y):
    mu_hat = np.mean(y)
    return mu_hat
# Bootstrapping:
S = 5000
sample = [ mle(y.sample(frac=1.0,replace=True)) for s in range(S)]

# Sampling distributions:
mu_hat_sample = [ row for row in sample]
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=mu_hat_sample,ax=axes[0],label='kde')
axes[0].set_title('Sampling Density, $\\hat{\\mu}$')
sns.ecdfplot(x=mu_hat_sample,ax=axes[1],label='ecdf')
axes[1].set_title('Sampling Distribution, $\\hat{\\mu}$')
plt.show()
```
